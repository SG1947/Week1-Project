```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a visual style
sns.set(style="whitegrid")
```

```python
# Load the CSV file
file_path = '/content/drive/MyDrive/Colab Notebooks/2016_Summary_Commodity.csv'
df = pd.read_csv(file_path)

# Preview the dataset
df.head()
```

```python
# Shape and info
print("Shape:", df.shape)
print("\nColumns:\n", df.columns)
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# Rename columns to remove leading/trailing spaces
df.columns = df.columns.str.strip()

# Drop rows with all NaNs (if any)
df.dropna(how='all', inplace=True)

# Display unique commodities
print("Unique Commodities:\n", df['Commodity Name'].unique())

# Display cleaned DataFrame head
df.head()
```

```python
# Summary statistics for numeric columns
df.describe()
```

```python
# Clean column names
df.columns = df.columns.str.strip()

# Group by Commodity and sum emission factors
top10_emissions = df.groupby('Commodity Name')['Supply Chain Emission Factors with Margins'] \
                    .sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=top10_emissions.values, y=top10_emissions.index, palette='magma')
plt.xlabel('Total Emission Factor (kg CO2e/unit)')
plt.ylabel('Commodity')
plt.title('Top 10 Commodities by Emission Factors (with Margins)')
plt.tight_layout()
plt.show()
```

```python
top10_names = df['Commodity Name'].value_counts().head(10).index
subset = df[df['Commodity Name'].isin(top10_names)]

plt.figure(figsize=(14,6))
sns.boxplot(data=subset, x='Commodity Name', y='Supply Chain Emission Factors with Margins', palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Distribution of Emission Factors for Top 10 Commodities')
plt.tight_layout()
plt.show()
```

```python
# Selecting DQ columns
dq_columns = [
    'DQ ReliabilityScore of Factors without Margins',
    'DQ TemporalCorrelation of Factors without Margins',
    'DQ GeographicalCorrelation of Factors without Margins',
    'DQ TechnologicalCorrelation of Factors without Margins',
    'DQ DataCollection of Factors without Margins'
]

plt.figure(figsize=(10, 6))
sns.heatmap(df[dq_columns].corr(), annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Correlation Heatmap of Data Quality Metrics')
plt.tight_layout()
plt.show()
```

```python
# Calculate mean DQ reliability score per commodity
top10_dq = df.groupby('Commodity Name')['DQ ReliabilityScore of Factors without Margins'] \
             .mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top10_dq.values, y=top10_dq.index, palette='Greens_r')
plt.xlabel('Average Reliability Score')
plt.title('Top 10 Commodities by DQ Reliability Score')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Supply Chain Emission Factors with Margins'], bins=40, kde=True, color='coral')
plt.title('Distribution of Emission Factors (with Margins)')
plt.xlabel('Emission Factor')
plt.ylabel('Frequency')
plt.xlim(0, df['Supply Chain Emission Factors with Margins'].quantile(0.95))
plt.tight_layout()
plt.show()
```
